### 3.
Function used: \( y = 0.001x^2 + \sin(0.03x) \).

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 1001)
y = 0.001 * x**2 + np.sin(0.03 * x)

plt.figure()
plt.plot(x, y)
plt.xlabel("x"); plt.ylabel("y"); plt.title("y = 0.001x² + sin(0.03x)")
plt.show()
```
